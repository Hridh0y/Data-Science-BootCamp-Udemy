```python
import pandas as pd
import numpy as np
```

```python
idf = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
idf.columns.values
```

```python
col = ['Reason Group 1', 'Reason Group 2',
       'Reason Group 3', 'Reason Group 4', 'Month', 'Day', 'Year', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']

df = idf[col]
df
```

```python
df.columns = ['Reason Group 1', 'Reason Group 2',
       'Reason Group 3', 'Reason Group 4','Month Value','Day of the Week', 'Year', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df
```

```python
targets = np.where(df['Absenteeism Time in Hours'] > df['Absenteeism Time in Hours'].median(), 1, 0)
targets
```

```python
df['Excessive Absenteeism'] = targets
df
```

```python
df
```

```python
df = df.drop(['Year','Date'],axis=1)
```

```python
df.iloc[:,:-2]
```

```python
unscaled_inputs = df.iloc[:,:-2]
```

# Standardizing and Splitting

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
scaled_inputs = absenteeism_scaler.fit_transform(unscaled_inputs)
```

```python
print(scaled_inputs)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs, targets)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = 0.8)
```

# Modeling
